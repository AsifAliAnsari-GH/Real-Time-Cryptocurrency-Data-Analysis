# Live-Running Excel Sheet

```python
import requests
import pandas as pd
import schedule
import time

import requests

url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&price_change_percentage=24h"

headers = {
    "accept": "application/json",
}


# Function to fetch and save live data to Excel
def fetch_and_update_excel():
    print("Fetching data...")

    # Fetch data from API
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        data = response.json()

        # Process the data into a DataFrame
        crypto_data = [
            {
                "Name": coin.get("name"),
                "Symbol": coin.get("symbol").upper(),
                "Current Price (USD)": coin.get("current_price"),
                "Market Cap": coin.get("market_cap"),
                "24h Volume": coin.get("total_volume"),
                "24h Change (%)": coin.get("price_change_percentage_24h")
            }
            for coin in data
        ]
        df = pd.DataFrame(crypto_data)

        # Save the DataFrame to Excel
        df.to_excel("Crypto_Live_Data.xlsx", index=False, engine="openpyxl")
        print("Excel file updated successfully!")
    else:
        print(f"Error fetching data: {response.status_code}")
        print(response.text)

# Initial data fetch
fetch_and_update_excel()

# Schedule the function to run every 5 minutes
schedule.every(5).minutes.do(fetch_and_update_excel)

# Keep the script running for continuous updates
while True:
    schedule.run_pending()
    time.sleep(1)
```
